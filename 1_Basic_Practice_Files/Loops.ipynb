```python
years_experience = {
    'Luke': 3,
    'Kelly': 6,
    'Ken': 4,
    'Alex': 3
}
```

```python
for person in years_experience:
    print(person)
```

```python
for value in years_experience.values():
    print(value)
```

```python
for item in years_experience.items():
    print(item)
```

```python
for key, value in years_experience.items():
    print(key, value)
```

```python
for key, value in years_experience.items():
    if value > 5:
        print(key, 'has', value, 'years experience')
```

```python
count = 1
while count < 5:
    print(count)
    count += 1
```

```python
years = 0
index = 0

years_list = list(years_experience.items())
while years < 5:
    key, years = years_list[index]
    if years >= 5:
        print(key,'has',years, 'years experience')
    index += 1

```
